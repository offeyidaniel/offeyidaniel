## Missing values

### Since not every country has surveys in every year and we dont have for example data for the year 2022 for the main columns: Health (Life Expectancy), Social Support, Economy (GDP per Capita), Generosity, Trust (Government Corruption), Freedom I will have to handel the missing values. I chose interpolation to do so.

```python
# finding missing values
df_happiness.isnull().sum()
```

```python
# Create a new dataframe
df_nan = df_happiness[df_happiness['Social Support'].isnull() == True]
```

```python
df_nan
```

### Observation: Most NaNs are connected to the year 2022, as we already suspected. 
### Solution: We will first sort out data by country and year. In That way we have the scores from the different countries in line and sorted by year. And then we use interpolation to close the "data gap". This form of inputation is for our analysis a better use, because the timeframe (2015-2023) is kind of to big and had a lot of different world crisises like corona to just use the mean to fill the gaps. 

```python
# Sort df_happniss by country & year
df_happiness.sort_values(['Country', 'Year'])
```

```python
# creating a subset 
df_happiness_sorted = df_happiness.sort_values(['Country', 'Year'])
```

```python
# interpolate the dataframe
df_happiness_sorted.interpolate()
```

```python
df_interpolated = df_happiness_sorted.interpolate()
```

```python
df_interpolated.head(10)
```

```python
# testing interpolated data for the year 2022
df_test= df_interpolated[df_interpolated['Social Support'].isnull() == True]
```

```python
df_test
```

```python
# finding missing values 
df_interpolated.isnull().sum()
```

```python
# using loc to extract Afghanistan subset
df_Af = df_interpolated.loc[df_interpolated['Country'] == "Afghanistan"]
```

```python
df_Af
```

### Observation: Since only Values from Afghanistan are missing, I will do another round of interpolation. But this time I am using the limit direction "backward", because by default the interpolation direction is forward and since Afghanistan was the first Country in line the first interpolation function couldn't use the data beforehand to create the filled-in data. 

```python
df_interpolated_2 = df_interpolated.interpolate(method ='linear', limit_direction ='backward')
```

```python
df_interpolated_2.isnull().sum()
```
